### 큐

선입선출: 삽입하면 왼편에 쌓인다

먼저 들어간게 먼저 삭제

```python
from collections import deque 

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7) #5를 삭제해야함
queue.popleft() #지금까지 5 , 2, 3, 7 (거꾸로 해서 pop left) 
queue.append(1)
queue.append(4)
queue.popleft()

print(queue)
queue.reverse()
print(queue)
```

```python
queue_1 = deque()
queue_1.append(5)
queue_1.appendleft(2)
queue_1.appendleft(3)
queue_1.appendleft(7)
queue_1.pop()
queue_1.appendleft(1)
queue_1.appendleft(4)
queue_1.pop()

print(queue_1)
```

```python
#반복문으로 구한 n!
def factorical_iterative(n):
    result = 1
    
    for i in range(1,n+1):
        result = result * i
        
    return result

def factorical_recursive(n):
    if n <=1 :
        return 1
    else:
        return n* factorical_recursive(n-1)
```

```python
factorical_iterative(5)
```

```python
factorical_recursive(5)
```

## DFS

Depth-Fist Search, 깊이 우선 탐색

그래프: 노드와 간선으로 표현, 노드와 간선으로 연결되있으면 두 노드는 인접하다

인접 행렬: 2차원 배열로 그래프의 연결 관계를 표현하는 방식
    
인접 리스트: 리스트로 그래프의 연결 관계를 표현하는 방식

연결되지 않은 노드끼리는 무한의 비용이라 작성

DFS는 특정한 경로로 탐색하다가 특정한 상황에서 최대한 깊숙이 들어가서 노드를 방문한 후 다시 돌아가 다른 경로로 탐색하는 알고리즘이다

```python
def dfs(graph,v,visited):
    #현재 노드를 방문 처리
    visited[v] = True
    print(v, end = '')
    
    for i in graph[v]:
        if not visited[i]:
            dfs(graph,i,visited)
#각 노드가 연결된 정보를 리스트 자료형으로 표현            
graph = [
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
[1,7]
]

visited = [False] * 9
```

```python

dfs(graph,1,visited)
```

정점: 위치, 간선: 연결선

인접한 노드끼리는 다른 색을 가졌다는 것

아이디어: 각 노드를 이웃 꼭지점과 다른 색으로 칠하며 같은 색의 꼭짓점이 서로 연결되어 있는지 확인

```python
import sys
sys.setrecursionlimit(10 ** 6)
#재귀를 사용해서 풀때 깊이 제한을 두는 코드

K = int(input())
#V: 정점의 개수, E: 간선의 개수
for _ in range(K):
    V, E = [int(x) for x in sys.stdin.readline().split()] #반복문으로 여러줄 입력받을 때 사용해야한다.(시간초과 문제)

    nodes = [[] for _ in range(V)] #노드 판별
    visited = [0] * V #각 노드가 방문한 정보 

    for _ in range(E): #각 줄에 인접한 두 정점의 번호: u,v
        u, v = [int(x) for x in sys.stdin.readline().split()]
        nodes[u - 1].append(v - 1)
        nodes[v - 1].append(u - 1)


    def search(node_idx):
        for v in nodes[node_idx]: #node_idx: 노드의 번호
            if visited[v] == 0: #1번 노드부터 탐색: 인덱싱 = 0 #0 기준으로 시작
                if visited[node_idx] == 1:
                    visited[v] = 2
                elif visited[node_idx] == 2:
                    visited[v] = 1 #인접한거에 다른거 채워줌
                s = search(v) #v는 위에서 처음 입력한 node_idx+1
                if not s:
                    return False
            elif visited[node_idx] == visited[v]: #색을 체크해서 같으ㅜ면 이분 그래프 위배
                return False
        return True

    result = True

    for i in range(0, V):
        if visited[i] == 0:
            visited[i] = 1
            result = search(i)
            if not result:
                break
       #for문 돌면서 출발 지점 설정한다. 이미 방문한 노드는 바운하지 않는다.

    print("YES") if result is True else print("NO")
```

```python
1번노드부터 탐색하므로 search(0) 호출한다. 

1번노드 방문했으므로 visited[0] = 1 #1이란 색으로 칠한다.

search함수에 node_idx = 0이면 nodes[0] 에는 [1,2] 존재. 
```
